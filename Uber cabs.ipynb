```python
#### Google map API call
apr=pd.read_csv('apr_data.csv', sep='\t', index_col=False)
may=pd.read_csv('may_data.csv', sep='\t', index_col=False)
```

```python
ls_coord=np.concatenate((apr["Coord"],may["Coord"]), axis=0)
```

```python
df=pd.DataFrame(list(set(ls_coord)))
df.columns=['cord']
df
```

```python
df['cord'].dtype
```

```python
### use google map api to read zipcode

from geopy.geocoders import Nominatim
import re
import time
import csv
geolocator = Nominatim(user_agent='uber taxi forecast')
zipcode = []
zipcode_all = []


for n in range(0,df.shape[0]):
    location = geolocator.reverse(df.iloc[n], timeout=50)
    #print(location.address)
    if location.address == None:
        zipcode = "NaN"
    else:
        num = str(re.findall(r"\D(\d{5})\D", location.address))
        #num = str(filter(str.isdigit, num))
        print(num)
        zipcode = num
        zipcode_all.append(zipcode[2:7])
    output = str(df.iloc[n]) + "," + zipcode
    print(output)
    time.sleep(1)
    with open(r"zipcode_output1.csv", 'a') as f:
        writer = csv.writer(f,delimiter='\t')
        writer.writerow()
```
